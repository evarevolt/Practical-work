```python
#Загрузка и индексирование PDF-документов
def load_and_index_documents():
    reader = SimpleDirectoryReader('./data')
    documents = reader.load_data()
    index = GPTVectorStoreIndex.from_documents(documents)
    return index
```

```python
#Индексирование
index = load_and_index_documents()
```

```python
#Промт для нейро-сотрудника, инструкции.
from llama_index.core.llms import ChatMessage, MessageRole
from llama_index.core.prompts import ChatPromptTemplate

system_message = (
            "Ты - психологический консультант, и твоя задача - помогать пользователям справляться с их проблемами и "
            "эмоциональными трудностями. Вот несколько ключевых моментов, которым нужно следовать в своей работе:\n\n"
            "1. Эмпатия и внимание: Выслушивай клиента внимательно, проявляй искреннюю заинтересованность и понимание его проблем.\n"
            "2. Без осуждения: Избегай осуждений и оценочных суждений, сосредоточься на поддержке и помощи.\n"
            "3. Активное слушание: Перефразируй и уточняй услышанное, чтобы убедиться в правильном понимании клиента.\n"
            "4. Поддержка и мотивация: Поощряй клиента к положительным изменениям, поддерживай его усилия и достижения.\n"
            "5. Практические советы: Предоставляй конкретные и практические советы на основе знаний из базы данных.\n"
            "6. Конфиденциальность: Никогда не разглашай личную информацию клиентов и соблюдай этические стандарты.\n"
            "7. Позитивный настрой: Стремись к тому, чтобы твои ответы были позитивными и обнадеживающими.\n"
            "8. Адаптация: Приспосабливайся к уникальным потребностям и стилю общения каждого клиента.\n"
            "9. Гибкость: Будь готов изменять подходы в зависимости от контекста и ситуации.\n"
            "10. Безопасность: Убедись, что ответы не содержат оскорблений, личной информации или других запрещённых данных."
        )

#Text QA Prompt
chat_text_qa_msgs = [
    ChatMessage(
        role=MessageRole.SYSTEM,
        content=system_message,
    ),
    ChatMessage(
        role=MessageRole.USER,
        content=(
            "Context information is below.\n"
            "---------------------\n"
            "{context_str}\n"
            "---------------------\n"
            "Given the context information and not prior knowledge, "
            "answer the query.\n"
            "Query: {query_str}\n"
            "Answer: "
        ),
    ),
]
text_qa_template = ChatPromptTemplate(chat_text_qa_msgs)

# Refine Prompt
chat_refine_msgs = [
    ChatMessage(
        role=MessageRole.SYSTEM,
        content=system_message,
    ),
    ChatMessage(
        role=MessageRole.USER,
        content=(
            "New Context: {context_msg}\n"
            "Query: {query_str}\n"
            "Original Answer: {existing_answer}\n"
            "New Answer: "
        ),
    ),
]
refine_template = ChatPromptTemplate(chat_refine_msgs)
```

```python
# Подготавливаем движок к индексу и задем ему вопрос
query_engine = index.as_query_engine(text_qa_template=text_qa_template, refine_template=refine_template, similarity_top_k=10, node_postprocessors=[cohere_rerank])
```

```python
#Обработка запроса и генерация ответа с системным сообщением

def handle_query(query, index):

    filtered_status=llamaguard_pack.run(query)
    if filtered_status == 'safe':
        full_query = f"{system_message}\n\nПользователь: {query}"
        response = index.query(full_query)
        response_status = llamaguard_pack.run(str(response))
        response = response if response_status == 'safe' else "Я не смогу вам помочь с этим. Задайте, пожалуйста, другой вопрос."
    else:
        response = "Вопрос не безопасный. Пожалуйста, введите другой вопрос."
    return response
```

```python
#Проверка работы с использованием фукции handle_query (LlamaGuardModeratorPack). Если вопрос будет не безопасный, то выдаст сообщение "Я не смогу вам помочь с этим. Задайте, пожалуйста, другой вопрос."
response = handle_query("Как перестать злиться в онлайн играх?", query_engine)
print(response)
```

```python
#Интерфейс для нейро-сотрудника на Gradio
import gradio as gr

#Основная функция для работы нейро-сотрудника с интерфейсом на Gradio
def main():
    global query_engine
    def chatbot_interface(query):
        return handle_query(query, query_engine)

    with gr.Blocks() as demo:
        gr.Markdown("# Нейро-сотрудник Алекс")
        gr.Markdown("Психологическая помощь и поддержка.")

        with gr.Row():
            with gr.Column():
                query = gr.Textbox(label="Введите ваш запрос", placeholder="Например, как справиться с тревогой?")
                submit_button = gr.Button("Отправить")

            with gr.Column():
                response = gr.Textbox(label="Алекс", interactive=False)

        submit_button.click(fn=chatbot_interface, inputs=query, outputs=response)

    demo.launch(share=True)

if __name__ == "__main__":
    main()
```
